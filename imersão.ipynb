## Aula 2 - Preparação e limpeza dos Dados

```python
df_substituir_remoto.isnull().sum()
```

```python
df_substituir_remoto['ano_trabalho'].unique()
```

```python
df_substituir_remoto[df_substituir_remoto.isnull().any(axis=1)]
```

```python
import numpy as np 

df_salarios = pd.DataFrame({
    'nome': ['João', 'Maria', 'Pedro', 'Ana', 'Val'], 
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(0))

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

display(df_salarios)
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'Nome': ['Ana', 'Maria', 'Pedro', 'João', 'Lucas'],
    'Cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém'],
})

df_cidades['Cidade_Preenchida'] = df_cidades['Cidade'].fillna('Não informado')

df_cidades
```

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Salário Médio por Nível de Senioridade (USD)',
             labels={'usd': 'Salário Médio (USD)', 'senioridade': 'Nível de Senioridade'})

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['Tipo_Trabalho', 'Contagem']

fig = px.pie(remoto_contagem,
             names='Tipo_Trabalho',
             values='Contagem',
             title='Contagem de Vagas por Tipo de Trabalho Remoto',
             labels={'Tipo_Trabalho': 'Tipo de Trabalho Remoto', 'Contagem': 'Contagem'})

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['Tipo_Trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='Tipo_Trabalho',
             values='quantidade',
             title='Contagem de Vagas por Tipo de Trabalho Remoto',
             labels={'Tipo_Trabalho': 'Tipo de Trabalho Remoto', 'quantidade': 'quantidade'},
             hole=0.5)

fig.update_traces(textinfo='percent+label')
fig.show()
```

### Gerar um gráfico de sálario médio por país do cargo 'Cientista de Dados'

```python
salario_medio_pais = df_limpo[df_limpo['cargo'] == 'Cientista de Dados'].groupby('residencia')['usd'].mean().reset_index().sort_values(by='usd', ascending=False)
salario_medio_pais.columns = ['País', 'Salário Médio (USD)']

fig = px.bar(salario_medio_pais,
             x='País',
             y='Salário Médio (USD)',
             title='Salário Médio de Cientistas de Dados por País (USD)',
             labels={'País': 'País', 'Salário Médio (USD)': 'Salário Médio (USD)'})

fig.show()
```
